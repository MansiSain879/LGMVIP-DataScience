```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/globalterrorismdb_0718dist.csv", encoding='latin1')
df.head()
```

```python
df.describe()
```

```python
df.dtypes
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
#Handling missing values

df['INT_LOG'].fillna(df['INT_LOG'].median(), inplace=True)
df['INT_IDEO'].fillna(df['INT_IDEO'].median(), inplace=True)
df['INT_MISC'].fillna(df['INT_MISC'].median(), inplace=True)
df['INT_ANY'].fillna(df['INT_ANY'].median(), inplace=True)
```

```python
df['country'].fillna(df['country'].mode()[0], inplace=True)
```

```python
df.columns
```

```python
#Checking again for null values
df.isnull().sum()
```

```python
plt.figure(figsize=(15, 8))
sns.countplot(x='region_txt', data=df)
plt.xlabel('Country')
plt.ylabel('Count')
plt.title('Distribution of Terrorist Attacks by Region')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
#Time Series Analysis
#Percentage Change and number of attacks by years on the same plot
# Calculating the number of attacks per year
yearly_attacks = df.groupby('iyear')['eventid'].count()

# Calculating the percentage change year over year
percentage_change = yearly_attacks.pct_change() * 100
plt.figure(figsize=(12, 8))

# Plotting the number of attacks
plt.plot(yearly_attacks.index, yearly_attacks, label='Number of Attacks', color='blue')

# Plotting the percentage change with secondary y-axis
plt.twinx()  # Creating a secondary y-axis
plt.plot(percentage_change.index, percentage_change, label='Year-over-Year % Change', color='red', linestyle='--')
plt.ylabel('% Change in Attacks', color='red')

plt.title('Number of Attacks Over the Years with Year-over-Year % Change')
plt.xlabel('Year')
plt.ylabel('Number of Attacks')
plt.legend(loc='upper left')
plt.grid(True)

plt.show()
```

```python
yearly_attacks = df.groupby('iyear')['eventid'].count()

# Calculating the percentage change year over year
percentage_change = yearly_attacks.pct_change() * 100

# Plot the percentage change
plt.figure(figsize=(12, 8))
plt.plot(percentage_change.index, percentage_change, label='Year-over-Year % Change', color='green', linestyle='--')
plt.title('Year-over-Year Percentage Change in Terrorist Attacks')
plt.xlabel('Year')
plt.ylabel('% Change in Attacks')
plt.grid(True)

# Highlighting the areas where attacks increased or decreased
plt.axhline(0, color='black', linestyle=':')
plt.fill_between(percentage_change.index, percentage_change, where=percentage_change > 0, color='green', alpha=0.3, label='Increase')
plt.fill_between(percentage_change.index, percentage_change, where=percentage_change < 0, color='red', alpha=0.3, label='Decrease')

plt.legend(loc='upper left')
plt.show()
```

```python
plt.figure(figsize=(6, 6))
extended_attacks = df['extended'].value_counts()
plt.pie(extended_attacks, labels=['Not Extended', 'Extended'], autopct='%1.1f%%', colors=['lightblue', 'orange'], startangle=90)
plt.title('Extended vs Non-Extended Attacks')
plt.show()
```

```python
plt.figure(figsize=(6, 6))
success_rate = df['success'].value_counts(normalize=True) * 100
plt.pie(success_rate, labels=['Successful', 'Unsuccessful'], autopct='%1.1f%%', colors=['green', 'red'], startangle=90)
plt.title('Percentage of Successful vs Unsuccessful Attacks')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='attacktype1_txt', data=df, order=df['attacktype1_txt'].value_counts().index)
plt.xticks(rotation=90)
plt.title('Distribution of Terrorist Attack Types')
plt.xlabel('Attack Type')
plt.ylabel('Count')
plt.show()
```

```python
#Analysing which weapon type is mostly used in terrorist attacks
plt.figure(figsize=(12, 6))
sns.countplot(x='weaptype1_txt', data=df, order=df['weaptype1_txt'].value_counts().index)
plt.xticks(rotation=90)
plt.title('Distribution of Weapon Types')
plt.xlabel('Weapon Type')
plt.ylabel('Count')
plt.show()
```

```python
#Correlation Matrix
plt.figure(figsize=(10, 8))
numeric_columns = ['nkill', 'nwound', 'extended', 'success']
sns.heatmap(df[numeric_columns].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix of Key Features')
plt.show()
```

```python
#Hostage-Situation Analysis
plt.figure(figsize=(6, 6))
hostage_situation = df['ishostkid'].value_counts()
plt.pie(hostage_situation, labels=['No Hostage', 'Hostage'], autopct='%1.1f%%', colors=['lightgray', 'lightcoral'], startangle=90)
plt.title('Hostage Situations')
plt.show()
```

```python
top_groups = df['gname'].value_counts().nlargest(10)
plt.figure(figsize=(12, 6))
sns.barplot(x=top_groups.index, y=top_groups.values, palette='Purples')
plt.xticks(rotation=90)
plt.title('Top 10 Terrorist Groups Involved in Attacks')
plt.xlabel('Group Name')
plt.ylabel('Number of Attacks')
plt.show()
```

```python
fatalities_per_year = df.groupby('iyear')['nkill'].sum()
plt.figure(figsize=(12, 6))
fatalities_per_year.plot(kind='line', color='red')
plt.title('Number of Fatalities Over the Years')
plt.xlabel('Year')
plt.ylabel('Number of Fatalities')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='targtype1_txt', data=df, order=df['targtype1_txt'].value_counts().index)
plt.xticks(rotation=90)
plt.title('Number of Attacks by Target Type')
plt.xlabel('Target Type')
plt.ylabel('Count')
plt.show()
```

```python
top_countries = df['country_txt'].value_counts().nlargest(10)
plt.figure(figsize=(12, 6))
sns.barplot(x=top_countries.index, y=top_countries.values)
plt.xticks(rotation=90)
plt.title('Top 10 Countries with the Most Terrorist Attacks')
plt.xlabel('Country')
plt.ylabel('Number of Attacks')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
df.groupby('attacktype1_txt')['nkill'].sum().sort_values(ascending=False).plot(kind='bar', color='red')
plt.title('Total Number of Kills by Attack Type')
plt.xlabel('Attack Type')
plt.ylabel('Total Number of Kills')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='iday', data=df, order=range(1, 32))
plt.title('Number of Attacks by Day of the Month')
plt.xlabel('Day of the Month')
plt.ylabel('Count')
plt.show()
```

```python
#Hostage incidents by Region
plt.figure(figsize=(12, 6))
df.groupby('region_txt')['ishostkid'].sum().sort_values(ascending=False).plot(kind='bar', color='orange')
plt.title('Number of Hostage Incidents by Region')
plt.xlabel('Region')
plt.ylabel('Number of Hostage Situations')
plt.xticks(rotation=90)
plt.show()
```

#Conclusion:
Increasing Trends: Based on time series analysis, we observed that terrorist attacks have generally increased over the years, with notable spikes from 1978-1980.

Geographic Hotspots: Countries like El Salvador, Peru, and Columbia have been significantly affected, with El Salvador experiencing the highest number of attacks. Additionally, South America  has consistently been a target of terrorist activity, particularly involving attacks by Explosives and Bombing.

Lethality and Attack Types: Armed Assault is the deadliest, resulting in the highest number of casualties. This suggest that certain types of attacks are far more lethal than others and require more focused security measures.

Terrorist Groups: A handful of groups, such as Group Shining Path(SL), are responsible for the majority of attacks, and they appear to rely heavily on specific weapons, such as Weapon Explosives, indicating a possible source of arms.

Target Group mostly attacked by terrorists are the people included in Business.

Hostage Situations: Hostage incidents are significantly concentrated in certain regions, such as South America, which poses additional humanitarian challenges that require both diplomatic and military interventions.
